```python
import requests
```

```python
#params로 url 만들기
headers={"User-Agent":"Mozilla/5.0"}
date="20210629"
page="2"
params={"date":date,"page":page} #이렇게도 url 연결가능.
naver_url="https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
r = requests.get(naver_url,params=params,headers=headers)
print(r.url)
```

```python
print(r.text)
```

```python
r.encoding #데이터 인코딩타입 확인가능, 한글꺠지면 r,encoding='utf-8' 이런식으로 지정가능
```

```python
r.content
```
