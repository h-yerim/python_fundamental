```python
for item2 in dl: 
    try:
        img = item2.find("dt",{"class":"photo"}).find("img")
        print("img= ",img["src"])
    except:
        print("no Image") #기사에 이미지가 없는 오류를 잡기위해 try-except문. 이미지 없으면 여기루 넘어온다
```

```python
#이미지 링크정보 가져오는 for문(완성스)
for no,item2 in enumerate(dl): 
    try:
        img = item2.find("dt",{"class":"photo"}).find("img")
        print("img= ",img["src"])
    except:
        print("no Image= ",no) #몇번째에서 에러가 걸렸는지 확인가능
```

```python
link =item3.find("dt",{"class":""}).find("a")
print("링크주소= ",link["href"])
print("제목 = ",link.text.replace("\t","").replace("\r\n","").strip() )
```

```python
#기사 링크 및 제목가져오기 뚝딱
for item in dl:
    try:
        link =item.find("dt",{"class":""}).find("a")
        print("링크주소= ",link["href"])
        print("제목 = ",link.text.replace("\t","").replace("\r\n","").strip() )
    except:
        print("No Content")
```

```python
#기사내용 가져오기 뚝딱스 (뒤에 근데 완벽하게 제거안될수도있음)
for item in dl:
    try:
        content=item.find("dd")
        print("내용= ",content.text.replace("\n","").split("…")[0])
    except:
        print("No Content")
```
