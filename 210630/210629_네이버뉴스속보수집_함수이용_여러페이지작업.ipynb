###  네이버 뉴스 속보 함수를 이용해서 여러페이지 작업
1. 날짜는 2021-06-28, 2021-06-29 2일만 가져온다.
 - 날짜별 page는 2개만
2. 4번의 page를 수집
3. 데이터프레임으로 만들기
 - ["이미지링크","기사링크","기사제목","기사내용","날짜","page"]
 - total 데이터프레임을 누적
4. total 데이터프레임 컬럼 이름 바꾸기
 - ["날짜","page","이미지링크","기사링크","기사제목","기사내용"]
 - total.shape로 확인

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
headers = { "User-Agent":"Mozilla/5.0"}  
```

```python
def get_page(url):
    html = requests.get(url, headers=headers)
    soup=BeautifulSoup(html.text,"html.parser")
    type06=soup.find("ul",{"class":"type06_headline"})
    dl=type06.find_all("dl")
    
    return dl
```

```python
def page_to_items(dl):
    total_news = []
    for no,item in enumerate(dl): 
        try:
            img = item.find("dt",{"class":"photo"}).find("img")
            link =item.find("dt",{"class":""}).find("a")
            content=item.find("dd")

            img_src=img["src"]
            link_href=link["href"]
            link_text=link.text.replace("\t","").replace("\r\n","").strip()
            content_text=content.text.replace("\n","").split("…")[0]
            total_news.append([img_src,link_href,link_text,content_text])

        except:
            print("No Data=  ",no)

    return total_news
```

```python
url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001&date=20210628&page=1"
get_dl=get_page(url)
total_news=page_to_items(get_dl)
#DataFrame만들기
news_df=pd.DataFrame(total_news,columns=["이미지링크","기사링크","기사제목","기사내용","날짜","page"])
pd.set_option("display.max.colwidth",50) 
     
news_df
```
