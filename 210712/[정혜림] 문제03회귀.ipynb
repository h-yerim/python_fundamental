```python
test=pd.read_excel('지역_위치별(주유소).xls')
df=test.copy()
df=df.dropna()
df
```

```python
label=df['고급휘발유']
tmp=df['휘발유']
tmp2=df['경유']
data=pd.concat([tmp,tmp2],axis=1)
```

```python
clf=LinearRegression()
clf.fit(data,label)
```
