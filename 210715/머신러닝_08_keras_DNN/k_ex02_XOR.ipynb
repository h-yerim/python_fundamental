```python
##############################################
# Modeling
##############################################
from tensorflow.keras import layers, models
from sklearn import preprocessing
import pandas as pd, numpy as np

def ANN_seq_func(Nin, Nh, Nout):
    model = models.Sequential()
    model.add(layers.Dense(8,input_shape=(Nin,),activation="tanh")) #relu써도됨
    model.add(layers.Dense(1,activation="sigmoid"))
    
    
    model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

    return model
```

```python
##############################################
#  Data 준비 
##############################################
import  numpy as np

def Data_func():
    X_train = np.array([[0,0],[1,0],[0,1],[1,1]])
    Y_train = np.array([[0],[1],[1],[0]] )
    
    print ("shape", X_train.shape , Y_train.shape) 
    return (X_train, Y_train    )
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##############################################
# Main   
##############################################
def main():
    Nin = 2
    Nh = 2
    Nout = 1
    model = ANN_seq_func(Nin, Nh, Nout)
    
    model.summary() #model이 어떤형태로 되어있는지(입력,출력,층은몇개) 출력해줌
    
    (X_train,   Y_train )  = Data_func() 
    print(Y_train) 
    history = model.fit(X_train, Y_train, epochs=1000, batch_size=5, \
                          verbose=1)

    history = history.history

    """Cost/Loss  변화 추이 그래프"""
    plt.plot(history['loss'])
#     plt.plot(history['val_loss'])
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Training', 'Verification'], loc=0)
    plt.show() 
    
    print(model.predict(X_train))
```

```python
if __name__ == '__main__':
    main()
```

  
tanh(x) 
tanh는 출력의 범위가 [-1, 1]이고 덕분에 sigmoid에 비해 높은 기울기를 가진다.
sigmoid를 사용해서 learning_rate를 높게 설정하는 것 보다 learning_rate를 낮추더라도 tanh를 사용하는게 훨씬 훈련이 빠르게 진행된다.
 sigmoid는 신경망의 층이 깊어지면 출력층에서 계산된 오차가 전파되면서 점점 감소하여 입력층에 가까운 가중치의 훈련이 잘 진행되지 않는 vanishing gradient 문제를 겪는다. tanh는 정도가 약하지만 여전히 문제가 존재하고, 이 문제를 해결하려면 ReLU계열의 함수가 필요하다.
