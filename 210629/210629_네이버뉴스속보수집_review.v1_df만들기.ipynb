```python
import requests
from bs4 import BeautifulSoup
headers = { "User-Agent":"Mozilla/5.0"}  
url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001" 

html = requests.get(url, headers=headers)
soup=BeautifulSoup(html.text,"html.parser")
type06=soup.find("ul",{"class":"type06_headline"})

dl=type06.find_all("dl")
total = []
for no,item in enumerate(dl): 
    try:
        img = item.find("dt",{"class":"photo"}).find("img")
        print("img= ",img["src"])
        a=img["src"]
        link =item.find("dt",{"class":""}).find("a")
        print("링크주소= ",link["href"])
        b=link["href"]
        print("제목 = ",link.text.replace("\t","").replace("\r\n","").strip() )
        c=link.text.replace("\t","").replace("\r\n","").strip()
        content=item.find("dd")
        print("내용= ",content.text.replace("\n","").split("…")[0])
        d=content.text.replace("\n","").split("…")[0]
        print("-"*50)
        total.append(a)
        total.append(b)
        total.append(c)
        total.append(d)
        
    except:
        print("No Data=  ",no)
```

```python
###선생님방법
import requests
from bs4 import BeautifulSoup
headers = { "User-Agent":"Mozilla/5.0"}  
url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001" 

html = requests.get(url, headers=headers)
soup=BeautifulSoup(html.text,"html.parser")
type06=soup.find("ul",{"class":"type06_headline"})

dl=type06.find_all("dl")
total_news = []
for no,item in enumerate(dl): 
    try:
        img = item.find("dt",{"class":"photo"}).find("img")
        link =item.find("dt",{"class":""}).find("a")
        content=item.find("dd")
        
        img_src=img["src"]
        link_href=link["href"]
        link_text=link.text.replace("\t","").replace("\r\n","").strip()
        content_text=content.text.replace("\n","").split("…")[0]
        total_news.append([img_src,link_href,link_text,content_text])

    except:
        print("No Data=  ",no)
```

```python
#DataFrame만들기
import pandas as pd
news_df=pd.DataFrame(total_news)
news_df.columns=["이미지링크","기사링크","기사제목","기사내용"]
pd.set_option("display.max.colwidth",50) #column 폭 키우는 방법
news_df.head()
```
