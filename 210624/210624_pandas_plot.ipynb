```python
iris.plot.scatter("petal_length","petal_width") #plot(kind="scatter") 해도됨
```

```python
def iris_scatter(x,y):
    iris.plot.scatter(x,y)
```

```python
iris_scatter("sepal_length","sepal_width")
```

```python
iris.plot(kind="box")
```

```python
bb = iris.species.value_counts() ;bb
```

```python
bb.plot(kind="pie")
```

```python
bb.plot(kind="pie", autopct="%.2f%%") #숫자 표시 (autopct)
```

```python
bb.plot(kind="pie", autopct="%.2f%%",shadow=True) #그림자
```

```python
bb.plot(kind="pie", autopct="%.2f%%",shadow=True,startangle=90) #시작점 각도
```

```python
bb.plot(kind="pie", autopct="%.2f%%",shadow=True,startangle=90, explode=(0,0,0.1)) #조각내줌
```
