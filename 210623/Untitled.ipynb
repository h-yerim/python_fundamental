6. 제목이 "Hamlet"인 영화는 몇개인가?

```python
count = 0 
for i in movies_title["title"]:
    if i == "Hamlet":
        count += 1
        
print("제목이 Hamlet인 영화는 ",count,"개입니다.")
```

7. 제목이 "Treasure Island"는 몇개인가? 제작년도 오름차순으로 나열

```python
count = 0
for i in movies_title["title"]:
    if i=="Treasure Island":
        count +=1
        
print("제목이 Treasure Island 인 영화는 ",count,"개")
```

8. 1950년에서 1959년 사이에 만들어진 영화는 몇개인가?

```python
count = 0
for i in movies_title["year"]:
    if (1950<=i)and(i<=1959):
            count += 1

print("1950년에서 1959년 사이에 만들어진 영화는",count,"개입니다.")
```
