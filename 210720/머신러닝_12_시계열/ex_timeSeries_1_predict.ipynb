```python
import numpy as np
import matplotlib.pyplot as plt
import ex_timeSeries_1_data as ex_t1_d
import ex_timeSeries_1_model as ex_t1_m

(X_train,Y_train) = ex_t1_d.make_data()    #샘플 데이터 load  

model= ex_t1_m.make_model() #RNN 모델 생성 ex_t1_m 파일을 model로 불러올거고
```

```python
# 학습 전 에측 결과 (예측이 잘 맞지않음을 알 수 있다.)

plt.plot(Y_train , 'ro-', label="target") # 원래 값
plt.plot(model.predict(X_train[:,:,:]), 'bs-', label="output") #학습전예측
plt.xlim(-0.5, 20.5)
plt.ylim(-1.1, 1.1)
plt.legend()
plt.title("Before training")
plt.show()
```

```python
# 학습
history = model.fit(X_train, Y_train, epochs=100, verbose=0)
```

```python
# 학습 후 loss
plt.plot(history.history["loss"])
plt.title("Loss")
plt.show()
```

```python
# 학습 후 예측 결과
plt.plot(Y_train, 'ro-', label="target")
plt.plot(model.predict(X_train[:,:,:]), 'bs-', label="output")
plt.xlim(-0.5, 20.5)
plt.ylim(-1.1, 1.1)
plt.legend()
plt.title("After training")
plt.show()
```
