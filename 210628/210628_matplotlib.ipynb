```python
def my_plotter(ax,data1,data2,param_dict):
    out = ax.plot(data1,data2,**param_dict) #매개변수로 받은 ax의 그림을 그리겠다. **는 키워드 딕셔너리. 몇개가들어올지모를때사용
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4,100)
fig, ax = plt.subplots(1,1)
my_plotter(ax,data1,data2,{'marker':'x'}) #marker를 x표시로 하겠다는 뜻.
```

```python
fig,(ax1,ax2) = plt.subplots(1,2)
my_plotter(ax1,data1,data2,{'marker':'x'})
my_plotter(ax2,data3,data4,{'marker':'o'})
plt.show()
```

```python
#위와 같은 결과 다른 방법
fig,ax = plt.subplots(1,2)
my_plotter(ax[0],data1,data2,{'marker':'x','color':'b'}) #color 옵션을 줄 수도 있다.
my_plotter(ax[1],data3,data4,{'marker':'o','color':'g'})
plt.show()
```

```python
fig,ax = plt.subplots(2,1)
my_plotter(ax[0],data1,data2,{'marker':'x','color':'b'}) 
my_plotter(ax[1],data3,data4,{'marker':'o','color':'g'})
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs') 파이 자르는거

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### 실습) 데이터는 tips
1. tips = sns.load_dataset("tips")
2. 요일별 분포(건수)를 pi 그래프로 그리기

```python
tips = sns.load_dataset("tips") ; tips
```

```python
index= tips["day"].value_counts().index
values= tips["day"].value_counts().values

fig1, ax1 = plt.subplots()
ax1.pie(values, explode=(0,0,0,0), labels=index, autopct='%d%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.show()
```

```python
def chart_pie(feature):
    index = tips[feature].value_counts().index
    values = tips[feature].value_counts().values
    
    fig1, ax1 = plt.subplots()
    ax1.pie(values, labels=index, autopct='%d%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  
    plt.show()
    
    return 
```

```python
chart_pie("size")
```

```python
chart_pie("time")
```

```python
#또다른방법
def chart_pie1(values,index):
    fig1, ax1 = plt.subplots()
    ax1.pie(values, labels=index, autopct='%d%%',
            shadow=True, startangle=90)
    ax1.axis('equal')  
    plt.show()
    
    return 
```

```python
index = tips["day"].value_counts().index
values = tips["day"].value_counts().values

chart_pie1(values,index)
```

### iris 데이터 scatter plot
1. x(petal_length),y(petal_width),c(sepal_length),s(sepal_width*100)

```python
iris=sns.load_dataset("iris")

data={ 'x':iris["petal_length"],
        'y':iris["petal_width"],
        'c':iris["sepal_length"],
        's':iris["sepal_width"] *100
     }

plt.scatter('x', 'y', c='c', s='s', data=data) 
plt.xlabel("petal_legnth")
plt.ylabel("petal_width")
plt.show()
```

```python
#선생님방법
iris=sns.load_dataset("iris")
x = iris["petal_length"]
y = iris["petal_width"]
fig,ax=plt.subplots()
ax.scatter(x,y,alpha=0.5,c=iris["sepal_length"],s=iris["sepal_width"]*100)

ax.set_xlabel("x축",fontsize=15)
ax.set_ylabel("y축",fontsize=15)
ax.set_title("Volume and percent change")
ax.grid(True)
```
