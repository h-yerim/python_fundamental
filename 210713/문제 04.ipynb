```python
import pandas as pd
shop=pd.read_csv('order.csv')
shop
```

```python
#KMeans
import mglearn
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

X=shop

agg=AgglomerativeClustering(n_clusters=5)
assignment=agg.fit_predict(X)

mglearn.discrete_scatter(X.iloc[:,0], X.iloc[:,1], assignment)
plt.legend(["cluster 0","cluster 1","cluster 2","cluster3","cluster4"],loc="best")
plt.xlabel("feature 0")
plt.ylabel("feature 1")
plt.show()
```

```python
#병합군집 => 시간이 좀 더 많이 걸린다. 그래서 보통 원형으로 분류가 되면 KMeans 많이씀
import mglearn
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc


#한글 폰트 등록
font_location = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)

X=shop

#######################################################################
#다섯 개의 클러스터 중심
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
assignments = kmeans.labels_  

mglearn.discrete_scatter(X.iloc[:,0],X.iloc[:,1],assignments) 

plt.legend(["클러스터1","클러스터2","클러스터3","클러스터4","클러스터5"])
plt.ylabel("특성1")
plt.xlabel("특성0")
plt.show()
```
