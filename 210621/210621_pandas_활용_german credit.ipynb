```python
german=pd.read_csv(url)
german_sample3=german[['Type of apartment','Sex & Marital Status','Credit Amount']]
german_sample3.head(3)
```

```python
for type1, group in german_sample3.groupby("Type of apartment"): # groupby에 의해 구한 type과 type에 해당하는 dataframe이 뜸(group)
    print("type= ",type1)
    print(group.head(3))
```

```python
for type1, group in german_sample3.groupby("Type of apartment"): 
    print("type= ",type1)
    print(group.shape) # dataframe 이기때문에 shape도 가능. type별 shape
```

```python
for (type1,sex),group in german_sample3.groupby(["Type of apartment","Sex & Marital Status"]):
    print((type1,sex)) #2가지 튜플을 가져다줌
    print(group.head(3))
```
